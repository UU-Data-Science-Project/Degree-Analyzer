The code blow is the scrapped data from website: http://online.wsj.com/public/resources/documents/info-Degrees_that_Pay_you_Back-sort.html where the html website was downloaded and than turned into a beautiful soup object to scrape. (Notes it does take a while to scrape). 

```python
#Data scrapping
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd

#Creating beautiful soup object 
Degrees_soup = BeautifulSoup(open("WSJ.com.html"), "html.parser")


#Heading Array 
heading = []
#n-1 length of heading array 
heading_len = np.zeros(7)
#Getting first entry in heading and clearing out the exccess string characters 
##heading.append(Degrees_soup.select(".header")[0].get_text())  
##heading[0] = heading[0].replace("\xa0\xa0\xa0","")
#Getting 2nd-8th entry title array 
for i, val in enumerate(heading_len):
    heading.append(Degrees_soup.select(".header2")[i].get_text().replace("\xa0\xa0\xa0",""))

#column to hold dataframe index 
column_index = [] 
  
#Next scraping for the rest of row elements
for i, val in enumerate(Degrees_soup.select(".col1")):
    column_index.append(val.get_text().replace("\xa0",""))
    
#size of column and rows 
col = 7
row = len(column_index)
#array to hold the rest of the values
column_info = np.zeros((row,col),dtype=float)

#indexers for filling row and columns 
counter = 0
tracker = 0
index = 0 

#While loop to fill in matrix
while tracker < row:
    while counter < col:
        column_info[tracker][counter]=float(Degrees_soup.select(".col2")[index].get_text().replace("\xa0","").replace("$","").replace(",",""))
        counter = counter + 1
        index = index + 1
    counter = 0
    tracker = tracker + 1
    
```

```python
df_Major = pd.DataFrame(column_info,index=column_index,columns=heading)
df_Major.to_csv("Project_Major_dataset.csv")
df_Major
```
